```python
polls_df=pd.read_csv("C:/Users/navya/OneDrive/Documents/Project-Election/dataset/president_primary_polls.csv")
```

```python
# Define the candidates you want to focus on
target_candidates = ['Donald Trump', 'Kamala Harris']

# Create a new column 'candidate_group' to categorize candidates
polls_df['candidate_group'] = polls_df['candidate_name'].apply(
    lambda x: x if x in target_candidates else 'Others'
)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoders
population_encoder = LabelEncoder()
subpopulation_encoder = LabelEncoder()

# Encode population and subpopulation
polls_df['population_encoded'] = population_encoder.fit_transform(polls_df['population'].fillna('Unknown'))
polls_df['subpopulation_encoded'] = subpopulation_encoder.fit_transform(polls_df['subpopulation'].fillna('Unknown'))
```

```python
polls_df['subpopulation_encoded']
```

```python
# Scatter plot for population vs subpopulation, colored by candidate_group
plt.figure(figsize=(10, 6))

# Create scatter plot for each candidate group
for candidate in polls_df['candidate_group'].unique():
    candidate_data = polls_df[polls_df['candidate_group'] == candidate]
    plt.scatter(
        candidate_data['population_encoded'], 
        candidate_data['subpopulation_encoded'], 
        label=candidate, 
        alpha=0.5
    )

# Add labels and title
plt.xlabel('Population')
plt.ylabel('Subpopulation')
plt.title('Comparison of Population and Subpopulation by Candidate Group')
plt.legend()
plt.xticks(ticks=range(len(population_encoder.classes_)), labels=population_encoder.classes_)
plt.yticks(ticks=range(len(subpopulation_encoder.classes_)), labels=subpopulation_encoder.classes_)
plt.grid(True)
plt.show()
```

```python
polling_data = polls_df[['candidate_group', 'answer', 'pct']]

# Create a bar plot comparing candidate_name and answer
plt.figure(figsize=(12, 6))
sns.barplot(x='candidate_group', y='pct', hue='answer', data=polling_data, ci=None)

# Add labels and title
plt.xlabel('Candidate Being Polled')
plt.ylabel('Polling Percentage (%)')
plt.title('Polling Percentage by Candidate and Respondent Answer')
plt.xticks(rotation=45)
plt.legend(title='Answer (Respondent Supported Candidate)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
target_answers = ['Trump', 'Harris']

# Create a new column 'candidate_group' to categorize candidates
polls_df['target_answer'] = polls_df['answer'].apply(
    lambda x: x if x in target_answers else 'Others'
)
```

```python
polling_data = polls_df[['candidate_group', 'target_answer', 'pct']]

# Create a bar plot comparing candidate_name and answer
plt.figure(figsize=(12, 6))
sns.barplot(x='candidate_group', y='pct', hue='target_answer', data=polling_data, ci=None)

# Add labels and title
plt.xlabel('Candidate Being Polled')
plt.ylabel('Polling Percentage (%)')
plt.title('Polling Percentage by Candidate and Respondent Answer')
plt.xticks(rotation=45)
plt.legend(title='Answer (Respondent Supported Candidate)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
size_population_data = polls_df[['sample_size', 'population', 'subpopulation']]

# Create a bar plot comparing candidate_name and answer
plt.figure(figsize=(12, 6))
sns.barplot(x='sample_size', y='population', hue='subpopulation', data=size_population_data, ci=None)

# Add labels and title
plt.xlabel('sample size')
plt.ylabel('population of voters')
 #(lv-Likely voters , rv-registered voters, a-adults, v-voters)
plt.title('No.of people opting for Democratic or Republic')
plt.xticks(rotation=45)
plt.legend(title='subpopulation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Create a bar plot to compare pollscore by pollster
plt.figure(figsize=(12, 6))
sns.barplot(x='pollster', y='pollscore', data=polls_df, ci=None)

# Add labels and title
plt.xlabel('Pollster')
plt.ylabel('Poll Score')
plt.title('Poll Score by Pollster')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by pollster and calculate the average pollscore
top_pollsters = polls_df.groupby('pollster')['pollscore'].mean().sort_values(ascending=False).head(10).index

# Filter the dataframe to include only the top 10 pollsters
df_top_pollsters = polls_df[polls_df['pollster'].isin(top_pollsters)]

# Create a bar plot for the top 10 pollsters
plt.figure(figsize=(12, 6))
sns.barplot(x='pollster', y='pollscore', data=df_top_pollsters, ci=None)

# Add labels and title
plt.xlabel('Pollster')
plt.ylabel('Poll Score')
plt.title('Top 10 Pollsters by Poll Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a box plot to compare pollscore by methodology
# Group by pollster and calculate the average pollscore
top_methodology = polls_df.groupby('methodology')['pollscore'].mean().sort_values(ascending=False).head(10).index

# Filter the dataframe to include only the top 10 pollsters
df_top_methodology = polls_df[polls_df['methodology'].isin(top_pollsters)]

plt.figure(figsize=(12, 6))
sns.barplot(x='methodology', y='pollscore', data=df_top_methodology)

# Add labels and title
plt.xlabel('Methodology')
plt.ylabel('Poll Score')
plt.title('Poll Score by Methodology')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a scatter plot to compare transparency_score vs pollscore
plt.figure(figsize=(10, 6))
sns.scatterplot(x='transparency_score', y='pollscore', hue='methodology', data=df_top_methodology)

# Add labels and title
plt.xlabel('Transparency Score')
plt.ylabel('Poll Score')
plt.title('Poll Score vs Transparency Score by Methodology')
plt.legend(title='methodology', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
